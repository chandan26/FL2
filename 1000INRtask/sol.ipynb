```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("us-states.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.describe().to_csv('describe.csv')
```

```python
plt.figure(figsize=(12,8))
#plt.xlabel("X axis label")
plt.ylabel("Deaths")
df.groupby(['state'])['deaths'].sum().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(12,8))
#plt.xlabel("X axis label")
plt.ylabel("Cases")
df.groupby(['state'])['cases'].sum().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(12,8))
#plt.xlabel("X axis label")
plt.ylabel("Deaths")
df.groupby(['fips'])['deaths'].sum().sort_values(ascending=False).plot(kind='bar')
```
